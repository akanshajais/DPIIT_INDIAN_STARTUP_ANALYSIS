# ANALYSIS AND VISUALIZATION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from datetime import datetime
import calendar
```

# ->  No. of startups Industry-wise

```python
#Number of startups , industry wise
industry = df_1.groupby('Industry').Count.sum().reset_index()
plt.figure(figsize=(15, 10))
sns.barplot(data=industry, x=industry.Industry, y=industry.Count)
plt.xticks(rotation=90)
plt.show()
```

# -> Top 10 Industries in startups

```python
industry.sort_values(by='Count', ascending=False).reset_index().head(10).plot.bar(x='Industry', y='Count', color = 'orange',
                                                                                  figsize=(15,10))
plt.title("Top 10 industries")
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15, 6))
df_year = df_1['Year'].value_counts().sort_index()
ax.bar(df_year.index, df_year, width=0.55,linewidth=0.7, color = 'blue')
for i in df_year.index:
    ax.annotate(f"{df_year[i]}",xy=(i, df_year[i] + 100),
                   va = 'center', ha='center')
ax.set_ylim(0, 1300)    
fig.text(0.1, 0.95, "Year Wise Growth of Startup's ", fontsize=15, fontweight='bold')    

ax.grid(axis='y', linestyle='-', alpha=0.4)  
```

# Insights Of Total Startups Year Wise (2016-2023)

```python
for i in years:
    df_1[df_1.Year == i].groupby(['Industry']).Count.sum().reset_index().sort_values(by='Count', ascending=False).plot.bar(x='Industry',color='brown',figsize=(15,10))
    plt.xlabel('Industry')
    plt.ylabel('Start ups count in year {}'.format((i)))
    plt.title("Industry wise start up counts in {}".format(i))
    plt.show()
```

```python
# AS TOP5 INDUSTRIES HAVE MAX NO. OF GROWTH IN STARTUPS ARE : IT SERVICES, HEALTHCARE & LIFESCIENCE, EDUCATION, AGRICULTURE. FOOD & BEVERAGES 
```

# State Wise Growth Of Top 5 Industries

```python
industries = ['IT Services', 'Healthcare & Lifesciences', 'Education', 'Agriculture', 'Food & Beverages']

for i in industries:
    top_industries = df_1[df_1.Industry==i]
    top_industries.groupby(['State']).Count.sum().reset_index().sort_values(by='Count', ascending=False).head().plot.bar(x='State',color = 'lightblue',figsize=(15,10))
    plt.xlabel('States')
    plt.ylabel('Start ups count in year {}'.format((i)))
    plt.title("State wise start up counts in {}".format(i))
    plt.show()
```

In all top5 Industries Maharastra record maximum number of startups.
